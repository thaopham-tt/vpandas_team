```python
import pandas as pd

import os
```

```python
# Tìm đường dẫn thư mục con và thư mục chính
current_dir = os.path.abspath(os.getcwd())
parrent_dir = os.path.dirname(current_dir)
base_dir = os.path.dirname(parrent_dir)
```

```python
meta_data = pd.read_json(base_dir + "\\Dữ liệu\\meta_Musical_Instruments.json", lines=True)
meta_data.head()
```

```python
meta_data.columns
```

```python
meta_data.drop(columns=["tech1", "fit", "also_buy", "tech2", "rank", "also_view", "similar_item", "date", "price",
                        "imageURL", "imageURLHighRes", "details", "main_cat", "description", "feature"], inplace=True)
meta_data.head()
```

```python
emptyCategoryIndex = meta_data[meta_data.category.str.len() < 1].index
emptyBrandIndex = meta_data[meta_data.brand.str.len() < 1].index
rowIndexToDrop = pd.Index.union(emptyBrandIndex, emptyCategoryIndex)

meta_data.drop(rowIndexToDrop, inplace=True)
meta_data.reset_index(drop=True, inplace=True)
meta_data.head()
```

```python
meta_data.shape
```

```python
meta_data = meta_data.assign(category=meta_data.category.apply(lambda x: x[-1]))
meta_data.category.head()
```

```python
meta_data.head()
```

```python
meta_data.to_csv("meta_data.csv", index=False)
```
